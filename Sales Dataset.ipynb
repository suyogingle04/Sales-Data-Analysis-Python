```python
import pandas as pd
```

```python
# Show Data
df = pd.read_csv(r"D:/Python/Superstore Sales Dataset/sales_data.csv", encoding='latin1')
```

```python
df
```

```python
# Show first 5 rowsprint(df.head())

print(df.head())
```

```python
# Check dataset info
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Handle missing values 
df =df.dropna()
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
# Convert date column (change column name if needed)
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
total_sales = df['SALES'].sum()
print("Total Sales:", total_sales)
```

```python
#TOP Product
top_products = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
#Sales by region
region_sales = df.groupby('COUNTRY')['SALES'].sum()
print(region_sales)
```

```python
# Extract month
df['Month'] = df['ORDERDATE'].dt.month

monthly_sales = df.groupby('Month')['SALES'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt
```

```python
top_products.plot(kind='bar')
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.ylabel('')
plt.show()
```
